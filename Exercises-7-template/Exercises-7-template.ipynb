Create the `sort_categories()` function, which takes a dictionary whose keys are movie categories and its values are their related frequencies. The function outputs a sorted list with tuples where the first element is the name of the category (string) and the second one is its frequency (integer). The list stores categories from most common to less common. Use the insertion sort algorithm to solve this exercise. In case of a tie, compare the categories by name.

Some code has been provided to test your implementation of the `sort_categories()` function in the following cell. 
The function `extract_all_categories()` gets a path to a file as input (string) and returns a dictionary. 
Use this code to verify that your implementation works as expected.
Additionally, the function `extract_all_categories()` gets a path to a file as input (string) and returns a dictionary.


**Example:**

*Input:*
```python
categories_dict = extract_all_categories('datasets/movie_titles_metadata.txt')
categories = sort_categories(categories_dict)
print(categories)
```

*Output:*
<pre>
[('drama', 40),
 ('thriller', 32),
 ('comedy', 21),
 ('action', 20),
 ('crime', 18),
 ('sci-fi', 18),
 ...]
</pre>

````python
# Run this cell to test your code and DO NOT modify the code in this cell

def extract_categories(line: str) -> List[str]:
    """ Returns the list of categories of the given 
        metadata line.
    """
    match = re.findall(r'\[(.+)\]', line)[0]
    categories = [c.strip().replace("'", "") for c in match.split(',')]
    return categories


def extract_all_categories(path: str) -> Dict[str, int]:
    """ Returns a dictionary where keys are categories and 
        values are the frequency of these categories in 
        the movies metadata file.
    """
    all_categories = dict()
    
    with open(path) as file:
        for line in file:
            categories = extract_categories(line)

            for category in categories:
                all_categories[category] = all_categories.get(category, 0) + 1
                
    return all_categories
````
