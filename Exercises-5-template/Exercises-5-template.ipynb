## 6. Modern Music Players

The record and cassette players are old-school music players.

For this exercise create the `ModernPlayer` class that inherits from the `MusicPlayer` class (created in exercise 5). 
This modern player will get all the music player attributes (`name`, `data_format`, and `state`), plus the `songs` attribute as parameters. The `songs` attribute is a list of strings representing the titles of the songs. It will also have the `current_song` attribute, which will keep track of the current song being played. It should start by pointing to the first item on the list.

Furthermore, the `ModernPlayer` class has the following two methods:
