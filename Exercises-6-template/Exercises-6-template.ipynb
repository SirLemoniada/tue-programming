## 4. Triangular Number Sequence

The triangular number sequence is a series of numbers that create equilateral triangles. 
Each subsequent number in the series adds a new row to the triangle as you can see below.

<div style="text-align:center">
	<img src="assets/triangular-sequence.png" alt="Triangular Number Sequence" width="500"/>
</div>

In the image, `n` represents the term in the sequence. 
With `n` we can compute the number of circles that are present in each triangle.
Thus, the $n^{th}$ term of the sequence is computed as $x_n = \frac{n(n+1)}{2}$.
For example, the $3^{rd}$ term of the sequence is computed as $x_3 = \frac{3(3+1)}{2} = \frac{12}{2} = 6$.
