## 6. Counting Prefixes

The function `histogram` creates a dictionary that stores the frequency of letters in a given `word`.

```Python
def histogram(word: str) -> Dict[str, int]:
    """Creates a dictionary that tracks the number of letters in a word.
    """
    
    dct = dict()
    for ch in word:
        if ch not in dct:
            dct[ch] = 1
        else:
            dct[ch] += 1
            
    return dct

```

Modify the function `histogram` to accept a list of strings `words` and an integer `n`.
The new function will return a dictionary corresponding to the histogram of the prefixes of `words`. The prefixes are obtained by selecting the first `n` characters of each word contained in `words`. 

**Note:** Assume that ```n >= 1``` and ```words``` is a non-empty list.

**Example:**  
*Input:*  
```Python

n = 2
```

*Output:*  
```Python
{ 'ho': 2, 'sh': 1, 're': 2, 'ro': 1}
```

````python
#// BEGIN_TODO [prefixes] Counting prefixes

# ===== =====> Replace this line by your code. <===== ===== #

#// END_TODO [prefixes]

words = ['hollow', 'home', 'show', 'rec', 'recital', 'rome']
n = 2
histogram(words, n)
````
